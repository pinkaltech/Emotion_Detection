```python
# Sentiment Analysis 
from textblob import TextBlob
```

```python
def get_sentiment(text):
    blob = TextBlob(text)
    
    sentiment = blob.sentiment.polarity
    print(sentiment)
    if sentiment > 0:
        result = "Positive"
    elif sentiment < 0:
        result = "Negative"
    else:
        result = "Neutral"
    return result
```

```python
get_sentiment('My name is pinkal , I feel so great and low but nice but not that much good')
get_sentiment('All my hopes were crushed by you')
get_sentiment('I completed all my goals this year')
get_sentiment('I really like the new design of your website!')
get_sentiment('The user interface is simple and does not necessitate extensive technical knowledge.')
get_sentiment('Working with large datasets is sometimes a struggle')
get_sentiment('The customer clearly isnt happy with the support theyve received,')
```

### Keyword Extraction
+ Extract most commonest words per class of emotion

```python
from collections import Counter
```

```python
def extract_keywords(text,num=50):
    tokens = [tok for tok in text.split()]
    most_common_tokens = Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
emotion_li = df['Emotion'].unique().tolist()
```

```python
emotion_li
```

```python
joy_list = df[df['Emotion'] == 'joy']['Clean_Text']
joy_list
```

```python
joy_docx = ' '.join(joy_list)
joy_docx
```

```python
# Extract Keywords
keyword_joy = extract_keywords(joy_docx)
```

```python
keyword_joy
```

```python
# Plot

def plot_most_common_words(mydict,emotion_name):
    plt.figure(figsize=(20,10))
    plt.title("Plot of {} Most Common Keywords".format(emotion_name))
    df_01 = pd.DataFrame(mydict.items(),columns = ['token','count'])
    sns.barplot(x='token',y='count',data=df_01)
    plt.xticks(rotation=45)
    plt.show()
    
```

```python
for emotion in emotion_li:
    emotion_list = df[df['Emotion'] == emotion]['Clean_Text']
    # print(emotion_list)
    emotion_text = " ".join(emotion_list)
    # print(emotion_text)
    emotion_keyword = extract_keywords(emotion_text)
    plot_most_common_words(emotion_keyword,emotion)
```

```python
def hypertuning_rscv(est,rfc_params,n_items,x,y):
    rdmsearch = RandomizedSearchCV(est,param_distributions=rfc_params,n_jobs=-1,n_iter=n_items,cv=9)
    rdmsearch.fit(x,y)
    ht_params = rdmsearch.best_params_ 
    ht_score = rdmsearch.best_score_ 
    return ht_params,ht_score
    
```
